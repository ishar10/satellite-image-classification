```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import imageio
img_width = 128
img_height = 128
batch_size = 30
```

```python


from scipy import ndimage, misc
import os
from csv import reader
from csv import writer
import csv


def main():
    outPath = 'C:/Users/HP/tensorflow/apni/'
    path = 'C:/Users/HP/tensorflow/train/'
    Adhered=0
    Concrete=0
    Plastic=0
    Steel=0
    Shingle=0
    Ballasted=0
    l=[]

    
    #print(l)
    l1=[]
    l2=[]
    with open('train_challenge.csv', 'r') as read_obj:
        csv_reader = reader(read_obj)
        for row in csv_reader:
            t=0
            if row[0]=="0":
#                 s1=str(0)+str(1)
#                 with open('train1_challenge.csv', 'a',newline='') as f_object:
#                     writer_object = writer(f_object)
#                     writer_object.writerow(s1)
                continue
            else:
                if row[1]=="Adhered":
                    if Adhered<715:
                        Adhered+=1
                        l1.append(str(row[0]))
                        t=1
                        l2.append( {'0':row[0], '1':row[1]})
                        
                    
                elif row[1]=="Concrete":
                    if Concrete<715:
                        Concrete+=1
                        l1.append(str(row[0]))
                        t=1
                        l2.append( {'0':row[0], '1':row[1]})
                
                elif row[1]=="Plastic & fabric":
                    if Plastic<715:
                        Plastic+=1
                        l1.append(str(row[0]))
                        t=1
                        l2.append( {'0':row[0], '1':row[1]})
                        
                
                elif row[1]=="Steel":
                    if Steel<715:
                        Steel+=1
                        l1.append(str(row[0]))
                        t=1
                        l2.append( {'0':row[0], '1':row[1]})
                    
                elif row[1]=="Shingle":
                    if Shingle<715:
                        Shingle+=1
                        l1.append(str(row[0]))
                        t=1
                        l2.append( {'0':row[0], '1':row[1]})
                    
                else:
                    if Ballasted<715:
                        Ballasted+=1
                        l1.append(str(row[0]))
                        t=1
                        l2.append( {'0':row[0], '1':row[1]})
#             if t==1:
#                 s=str(row[0])+str(row[1])
#                 with open('train1_challenge.csv', 'a',newline='') as f_object:
#                     writer_object = writer(f_object)
#                     writer_object.writerow(s)
    
    headers = ['0', '1']

    with open('train1_challenge.csv', 'w',newline='') as outfile:
        mywriter = csv.DictWriter(outfile, fieldnames=headers)
        mywriter.writeheader()

        for d in l2:
            mywriter.writerow(d)                    

    # iterate through the names of contents of the folder
    for image_path in os.listdir(path):
        if str(image_path) in l1:
            input_path = os.path.join(path, image_path)
            image = imageio.imread(input_path)
            fullpath = os.path.join(outPath,image_path)
            imageio.imwrite(fullpath, image)

if __name__ == '__main__':
    main()
```

```python
datagen = ImageDataGenerator(rescale=1/255.0,validation_split=0.15)
train1=pd.read_csv('train1_challenge.csv')
```

```python
train_generator = datagen.flow_from_dataframe(train1,directory='C:/Users/HP/tensorflow/apni/',
                                              target_size = (img_width, img_height),
                                              x_col='0', y_col='1',
                                              class_mode = 'categorical',
                                              class_color='greyscale',
                                              batch_size=batch_size,shuffle=True,subset="training")
```

```python
from scipy.stats import itemfreq
train_generator.classes
itemfreq(train_generator.classes)
```

```python
val_generator = datagen.flow_from_dataframe(train1,directory='C:/Users/HP/tensorflow/apni/',
                                              target_size = (img_width, img_height),
                                              x_col='0', y_col='1',
                                              class_mode = 'categorical',
                                            class_color='greyscale',
                                              batch_size=batch_size,shuffle=True,subset="validation")
```

```python
itemfreq(val_generator.classes)
```

```python
train_generator.labels
```

```python
model = Sequential()
model.add(Conv2D(16, (3,3), activation='relu', input_shape = (img_width,img_height,3)))
#model.add(Dropout(0.2))
#model.add(Conv2D(32, (3,3), activation='relu'))
model.add(MaxPool2D())
model.add(Flatten())
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(6, activation='softmax'))
```

```python
model.summary()
```

```python

model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history=model.fit_generator(generator=train_generator,steps_per_epoch=len(train_generator),epochs=10)
```
